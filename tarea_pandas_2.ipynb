```python
import numpy as np
import pandas as pd
```

```python
url ='https://raw.githubusercontent.com/Sinchiguano/chicagoUniversity/main/cta_Ridership.csv'
rs=pd.read_csv(url)
print(rs.shape)

rs.head
```

```python
rs['rides'].describe()
```

```python
print('Min: ',rs['rides'].min())
print('Max: ',rs['rides'].max())
```

```python
# Convertir la columna 'date' a tipo de dato 'DateTime'
rs['date'] = pd.to_datetime(rs['date'])

# Filtrar los datos desde el 1 de enero de 2001 hasta el 31 de diciembre de 2019
start_date = pd.to_datetime('2001-01-01')
end_date = pd.to_datetime('2019-12-31')
filtered_rider = rs[(rs['date'] >= start_date) & (rs['date'] <= end_date)]
filtered_rider.set_index('date', inplace=True)
print(filtered_rider)
```

```python
rs = filtered_rider.sort_values('date')
print(rs)
```

```python
Garfield_Dan_Ryan = rs[rs['stationname'] == 'Garfield-Dan Ryan']
Garfield_South_Elevated = rs[rs['stationname'] == 'Garfield-South Elevated']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(11, 4)})
Garfield_Dan_Ryan['rides'].plot(linewidth=0.5)
plt.show()

sns.set(rc={'figure.figsize':(11, 4)})
Garfield_South_Elevated['rides'].plot(linewidth=0.5)
plt.show()
```
